```python
import pandas as pd
import requests
import time
import numpy as np
from bs4 import BeautifulSoup

test_url = 'https://www.basketball-reference.com/leagues/NBA_2025_per_game.html#per_game_stats::pts_per_g'
data = requests.get(test_url)
soup = BeautifulSoup(data.text, 'html.parser')
```

```python
r = soup.find('tr')
```

```python
soup.find(id = 'Totals')
```

```python
all_elements_with_id = soup.find_all(id=True)  # Find all elements that have an 'id' attribute
all_ids = [element['id'] for element in all_elements_with_id]  # Extract the 'id' values

# Display all unique IDs found
print(set(all_ids))
```

```python
soup.find(id = 'all_per_game_stats')
```

```python
# Assuming 'soup' is already created from the previous steps
import pandas as pd

# Find the table by its ID
table = soup.find('table', id='per_game_stats')

# Extract headers from the table
headers = [th.get_text(strip=True) for th in table.find('thead').find_all('th')]

# Extract rows from the table
table_rows = table.find('tbody').find_all('tr')

# Extract data from each row
data = []
for tr in table_rows:
    cells = tr.find_all(['td', 'th'])  # This includes 'th' in case there are header rows inside the body
    row_data = [cell.text.strip() for cell in cells]
    data.append(row_data)

# Create a DataFrame
df = pd.DataFrame(data, columns=headers)

# Display the DataFrame
df
```
