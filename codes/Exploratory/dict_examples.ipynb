```python
import requests
from bs4 import BeautifulSoup

url = "https://sports.yahoo.com/nba/odds/"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# Extract relevant information (you'll need to inspect the webpage's structure)
odds_data = []
for game in soup.find_all("div", class_="some-class"):  # Adjust based on the actual structure
    teams = game.find_all("span", class_="team-name")
    odds = game.find_all("span", class_="odds-value")
    if teams and odds:
        odds_data.append({
            "team_1": teams[0].text,
            "team_2": teams[1].text,
            "odds_1": odds[0].text,
            "odds_2": odds[1].text,
        })

df = pd.DataFrame(odds_data)
df
```
